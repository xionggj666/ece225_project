2. https://www.ons.gov.uk/peoplepopulationandcommunity/crimeandjustice/datasets/crimeinenglandandwalesquarterlydatatables


```python
import pandas as pd
```

```python
def data_load(files):
    # quarters = [1, 2, 3, 4]
    quarter_df = pd.DataFrame(columns=['quarter', 'count'])
    counts = []
    for index,file in enumerate(files):
        df = pd.read_excel(file, sheet_name="Current QT1-PRC by Quarter")
        filtered_row = df.loc[df.iloc[:, 0] == "TOTAL RECORDED CRIME - ALL OFFENCES INCLUDING FRAUD"]
        values = filtered_row.iloc[0, 1:9].tolist()
        if index < 2:
            values = [round(value * 1000) for value in values]
        else:
            values = [round(value) for value in values]
        counts = counts + values
    for index, count in enumerate(counts):
        quarter = (index + 1) % 4  
        if quarter == 0:  
            quarter = 4
        quarter_df = quarter_df.append({'quarter': quarter, 'count': count}, ignore_index=True)
    quarter_df['quarter'] = pd.to_numeric(quarter_df['quarter'], errors='coerce')
    quarter_df['count'] = pd.to_numeric(quarter_df['count'], errors='coerce')
    return quarter_df
    
```

```python
# quarters = ["Q1", "Q2", "Q3", "Q4"]
# quarter_df = pd.DataFrame(columns=quarters)
files = ['quarterlydatatableyearendingdecember2015.xls', 'quarterlydatatableyearendingdecember2017.xls','quarterlydatatablesyedec19.xlsx','quarterlydatatablesdec21.xlsx','quarterlydatatablesyedec23.xlsx']
files2 = ['quarterlydatatablesyedec19.xlsx','quarterlydatatablesdec21.xlsx','quarterlydatatablesyedec23.xlsx']
```

```python
quarter_df = data_load(files)
# quarter_df = data_load(files2, quarter_df)
```
